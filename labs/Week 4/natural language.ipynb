```python
def is_a_real_word(x):
   return x not in "..[]:"
```

```python
clean_up_hamlet = [
    x.lower()
        for x in nltk.corpus.gutenberg.words("shakespeare-hamlet.txt")
        if is_a_real_word(x)
]
```

```python
distinct_words = set(clean_up_hamlet)
```

```python
distinct_words
```

```python
len(distinct_words)
```

```python
len(clean_up_hamlet)
```

```python
lexical_diversity_of_Hamlet = float( len(distinct_words)) / float(len(clean_up_hamlet))
```

```python
lexical_diversity_of_Hamlet
```

```python
nltk.corpus.wordnet.synsets('car')[0].definition() 
#will provide definition of first one
```

```python
nltk.corpus.wordnet.synsets('car')[0].lemma_names()
#will provide 
```

```python
car = nltk.corpus.wordnet.synsets('car')[0]
car.hypernms()
#will define a hierachy of definition eg self propelled conveyance
```

```python
bus = nltk.corpus.wordnet.synsets('bus')[3]
bus.hypernms()
#will provide the 4th hypernm of bus
```

```python
jeep = nltk.corpus.wordnet.synsets('jeep')[0]
jeep.lowest_common_hypernms(bus)
#will provide the nearest commonality
```

```python
banana = nltk.corpus.wordnet.synsets('banana')
banana[1].definition()
#gives 2nd definition of banana
```

```python
jeep.lowest_common_hypernms(banana)
#will provide the nearest commonality to a banana
```
