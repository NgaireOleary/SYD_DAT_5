```python
import pandas
import sklearn.cluster
import matplotlib.pyplot
import sklearn.datasets
```

```python
b = sklearn.datasets.load_boston()
```

```python
b.keys()
```

```python
b['feature_names']
```

```python
import pandas
Boston = pandas.DataFrame(data=b['data'], columns=b['feature_names'])
Boston['price'] = b['target']
```

```python
import sklearn.ensemble
```

```python
bag = sklearn.ensemble.BaggingRegressor(max_features = 3,
                                       n_estimators = 25,
                                       max_samples = 100,
                                       n_jobs = 1)
```

```python
boston_predictors = Boston.drop ('price', axis = 1)
boston_predictors
```

```python
bag.fit(boston_predictors, Boston.price)
```

```python
bag.estimators_
```

```python
zip(bag.predict(boston_predictors), Boston.price)
```

```python
%matplotlib inline
```

```python
import matplotlib.pyplot
matplotlib.pyplot.scatter(Boston.price, bag.predict(boston_predictors))
```

```python
rdf = sklearn.ensemble.RandomForestRegressor(n_estimators = 100)
```

```python
rdf.fit(boston_predictors, Boston.price)
```

```python
matplotlib.pyplot.scatter(Boston.price, rdf.predict(boston_predictors))
```

```python
rdf.estimators_
```

```python
zip(rdf.feature_importances_, boston_predictors.columns)
```
